Sales Analysis in PANDAS

```python
import pandas as pd
import os
```

Merge months data in to single file

```python
# example of reading a csv file
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
df
```

```python
# creating a file list which contains all files in Sales_Data
files = [file for file in os.listdir("./Sales_Data")]
# creating a DataFrame
all_month_data = pd.DataFrame()
# reading all csv files in folder/ directory
for file in files:
    df = pd.read_csv("./Sales_Data/" + file)
    all_month_data = pd.concat([all_month_data,df])

all_month_data
```

```python
all_month_data.to_csv("all_data.csv", index=False)
```

Remove NULL or NAN values

```python
nan_df = all_month_data[all_month_data.isna().any(axis=1)]
nan_df
```

```python
na = all_month_data.isna()
na
```

```python
all_month_data = all_month_data.dropna(how="any")
all_month_data
```

Add Month column

```python
# removing the OR instead of 04 in Month column
all_month_data = all_month_data[all_month_data["Order Date"].str[0:2] != "Or"]
all_month_data
```

```python
all_month_data["Month"] = all_month_data["Order Date"].str[0:2]
all_month_data["Month"] = all_month_data["Month"].astype("int32")
all_month_data
```

Add Sales

```python
# convert prices and quantity into int
all_month_data["Price Each"] = pd.to_numeric(all_month_data["Price Each"])
all_month_data["Quantity Ordered"] = pd.to_numeric(all_month_data["Quantity Ordered"])
```

```python
all_month_data["Sales"] = all_month_data["Price Each"] * all_month_data["Quantity Ordered"]
all_month_data
```

```python
monthly_sales = pd.DataFrame()
monthly_sales = all_month_data.groupby("Month").sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months,monthly_sales["Sales"])
plt.xticks(months)
plt.xlabel("Months")
plt.ylabel("Sales in USD")
```

What City has Highest Number of Sales

```python
all_month_data["City"] = all_month_data["Purchase Address"].apply(lambda x:x.split(",")[1]) 
# .apply() takes each datapoint  of Purchase add. 
# each datapoint is represented by x.
# split func splits any string x by the ,'s present in x.		
all_month_data
```

```python
# get city and state
def get_city(address):
    return address.split(",")[1]

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_month_data["City"] = all_month_data["Purchase Address"].apply(lambda x : get_city(x) + " " + get_state(x))

all_month_data
```

```python
city_sales = all_month_data.groupby("City").sum()
city_sales
```

```python
# cities = all_month_data["City"].unique()   
cities = [city for city, df in all_month_data.groupby("City")]

import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(cities,city_sales["Sales"])
plt.xticks(cities, rotation = 'vertical', size = 8)
plt.xlabel("Cities")
plt.ylabel("Sales in USD")
```
